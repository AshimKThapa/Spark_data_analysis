```python
import findspark
findspark.init()

from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("FlightAnalysis_Test") \
    .master("local[*]") \
    .config("spark.driver.memory", "4g") \
    .getOrCreate()

print("✅ Spark Version:", spark.version)
print("✅ Java Version Used:", spark.sparkContext.getConf().get("spark.driver.extraJavaOptions"))
spark.stop()
```
